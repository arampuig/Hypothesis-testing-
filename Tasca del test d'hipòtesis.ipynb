```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
import statistics
from scipy import stats
import scipy.stats

pd.set_option('display.max_columns', 85)
pd.set_option('display.max_rows', 85)
```

```python
df = pd.read_csv(r'mlbBat10.txt', delimiter = "\t")

df
```

```python
# Farem un 'Normality Test' per començar, amb la variable de partits jugats per jugador ('G'), 
# per saber si té una distribució Gaussiana.

# Assumcions
# Les observacions en cada mostra són independents i idènticament distribuides

# Interpretació

# H0: la mostra té una distribució normal
# H1: la mostra no té una distribució normal

from scipy.stats import shapiro
data = df['G'].sample(10)
stat, p = shapiro(data)
print('stat=%.3f, p=%.3f' %(stat,p))
if p > 0.05:
    print('Probably Normal')
else: 
    print('Probably not Normal')
```

```python
for i in df.columns[3:]:
    data = df[i].sample(10)
    stat, p = shapiro(data)
    print('stat=%.3f, p=%.3f' %(stat,p))
    if p > 0.05:
        print('The sample of {} is Probably Normal'.format(i))
    else: 
        print('The sample of {} is Probably not Normal'.format(i))
```

```python
# Nivell 2
# - Exercici 2
# Continua amb el conjunt de dades de tema esportiu que t'agradi i selecciona dos atributs del conjunt de dades.
#Calcula el p-valor i digues si rebutja la hipòtesi nul·la agafant un alfa de 5%.
```

```python
# Farem un Pearson's correlation test per comprovar la correlació entre dues variables. 
# Agafarem el numerode 'runs i el numero de 'hits'

# Assumptions

# Observations in each sample are independent and identically distributed (iid).
# Observations in each sample are normally distributed.
# Observations in each sample have the same variance.
# Interpretation

# H0: the two samples are independent.
# H1: there is a dependency between the samples.

from scipy.stats import pearsonr
data1= df['R'].sample(10)
data2= df['H'].sample(10)
stat, p = pearsonr(data1, data2)
print('stat=%.3f, p=%.3f' %(stat, p))
if p > 0.05:
    print('Probably independent')
else:
    print('Probably dependent')
```

```python
# Analysis of Variance Test (ANOVA)
# Tests whether the means of two or more independent samples are significantly different.

# Assumptions

# Observations in each sample are independent and identically distributed (iid).
# Observations in each sample are normally distributed.
# Observations in each sample have the same variance.
# Interpretation

# H0: the means of the samples are equal.
# H1: one or more of the means of the samples are unequal.
```

```python
# Example of the Analysis of Variance Test
from scipy.stats import f_oneway
data1 = df['R'].sample(10)
data2 = df['H'].sample(10)
data3 = df['AB'].sample(10)
stat, p = f_oneway(data1, data2, data3)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('Probably the same distribution')
else:
    print('Probably different distributions')
```
